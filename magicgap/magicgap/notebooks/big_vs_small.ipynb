```python
from magicgap import *
```

```python
d_b = 32
qudit_big_se = avg_magic_analytic(d_b)
nqubit_big_se = avg_magic_analytic(d_b, nqubits=True)
d_s_range = np.arange(2, d_b)
qudit_se_gaps = [qudit_big_se - avg_magic_analytic(d_s) for d_s in d_s_range]
qudit_se_gaps2 = [qudit_big_se - avg_magic_analytic(d_s, nqubits=is_pow2(d_s)) for d_s in d_s_range]
nqubit_se_gaps = [nqubit_big_se - avg_magic_analytic(d_s) for d_s in d_s_range]
nqubit_se_gaps2 = [nqubit_big_se - avg_magic_analytic(d_s, nqubits=is_pow2(d_s)) for d_s in d_s_range]
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(1)
ax.set_title("Average SE Gap, averaged over subspaces")
ax.set_xlabel("$d_S$")
ax.set_ylabel("Avg SE gap")
ax.grid(True)
ax.plot(d_s_range, qudit_se_gaps, label="qudit/qudit")
ax.plot(d_s_range, nqubit_se_gaps, label="n-qubit/qudit")
ax.plot(d_s_range, qudit_se_gaps2, linestyle="--", label="qudit/n-qubit")
ax.plot(d_s_range, nqubit_se_gaps2, linestyle=":", label="n-qubit/n-qubit")
ax.axhline(0, color="black", linestyle="-.")
ax.legend()
plt.tight_layout()
plt.savefig("img/big_vs_small")
plt.show()
```
